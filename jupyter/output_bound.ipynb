```python
from time import time

pbin = '/home/loitg/ssd/data/temp3/protobuf_unify_old'
imagesin = '/home/loitg/workspace/dewarp/hinhphang1/'

imagestemp = '/home/loitg/workspace/backbone/imagestemp/'

for i, fn in enumerate(os.listdir(imagesin)[:10]):
    if fn not in ['650000000945170.ae71c488-564f-435d-92a6-bbd88725b14d.636613201821177483.jpg']:
        continue
    print(fn)
    
    image = cv2.imread(os.path.join(imagesin, fn))
    r = 1.0
    image = cv2.resize(image, None,fx = r, fy = r)

    pb_response = readPb(os.path.join(pbin, change_ext(fn, 'pb', method='replace')))
    symb_bounds = get_document_bounds(pb_response, FeatureType.SYMBOL)
    word_bounds = get_document_bounds(pb_response, FeatureType.WORD)
    boxes = [bound2points(bound) for bound in word_bounds]
    
    mask_symb = np.zeros(image.shape[:2], dtype=np.uint8)
    for bound in symb_bounds:
        cv2.fillPoly(mask_symb, np.expand_dims(np.array(bound2points(bound), dtype=np.int32), axis=0), 1)
        
    mask_word = np.zeros(image.shape[:2], dtype=np.uint8)
    for box in boxes:
        cv2.fillPoly(mask_word, np.expand_dims(np.array(box, dtype=np.int32), axis=0), 1)
        
    
    rects = findRect(image, boxes)

    cv2.imwrite(os.path.join(imagestemp, '%s_symb.jpg' % (fn)), mask_symb*255)
    cv2.imwrite(os.path.join(imagestemp, '%s_word.jpg' % (fn)), mask_word*255)
    
#     image_bounds = image.copy()
#     draw_bounds(image_bounds, word_bounds)
#     cv2.imwrite(os.path.join(imagestemp, '%s_word.jpg' % (fn)), mask_word)

#     shutil.copy(os.path.join(imagesin, fn), imagestemp2)
```
